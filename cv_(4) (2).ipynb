```python
import pandas as pd
import numpy as np

def group_similar_majors(major):
    major = str(major).lower()

    if any(keyword in major for keyword in ['engineering', 'technology', 'computer', 'software', 'it']):
        return 'Engineering & Technology'
    elif any(keyword in major for keyword in ['business', 'commerce', 'mba', 'management']):
        return 'Business & Management'
    elif 'science' in major:
        return 'General Science'
    else:
        return 'Other'

df['unified_major'] = df['major_field_of_studies'].apply(group_similar_majors)

print(df['unified_major'].value_counts())
print("-" * 50)

top_unified_major = df['unified_major'].value_counts().idxmax()
df_filtered = df[df['unified_major'] == top_unified_major].copy()

print(f"✅ تم فلترة البيانات بناءً على المجال الموحد: {top_unified_major}")
print(f"📊 حجم البيانات الجاهزة للنموذج الآن: {df_filtered.shape}")
```

```python
import pandas as pd

df = df_filtered.copy()

columns_to_keep = [
    'major_field_of_studies',
    'skills',
    'start_dates',
    'end_dates',
    'educationaL_requirements',
    'experiencere_requirement',
    'skills_required',
    'matched_score'
]

df = df[columns_to_keep]

print("✅ تمت تصفية الأعمدة بنجاح.")
display(df.head(15))
```

```python
import ast

def to_list(x):
    """دالة لتحويل النص الذي يبدو كقائمة إلى قائمة برمجية فعلية"""
    if pd.isna(x):
        return []
    if isinstance(x, list):
        return x
    s = str(x).strip()
    try:
        v = ast.literal_eval(s)  # يقرأ نصوص مثل "['Python', 'Java']"
        return v if isinstance(v, list) else [v]
    except:
        return [s] if s else []

def join_list(x, sep=" | "):
    items = [str(i).strip() for i in to_list(x) if str(i).strip() and str(i).strip().lower() != "nan"]
    return sep.join(items)

# تطبيق الدوال لإنشاء أعمدة جديدة نظيفة وجاهزة للنموذج
df["skills_preview"] = df["skills"].apply(lambda x: join_list(x, sep=" | "))
df["edu_preview"] = df["major_field_of_studies"].apply(lambda x: join_list(x, sep=" | "))

print("✅ تم تحويل النصوص بنجاح.")
display(df[['skills_preview', 'edu_preview']].head(15))
```

```python
import re

def parse_date_any(s):
    """دالة لقراءة التاريخ وتحويل كلمات مثل Present إلى تاريخ اليوم"""
    if not s or pd.isna(s):
        return pd.NaT
    s = str(s).strip()

    if re.search(r"(till\s*date|present|current)", s, flags=re.I):
        return pd.Timestamp.today().normalize()

    return pd.to_datetime(s, errors="coerce")

def years_experience_from_lists(starts_cell, ends_cell):
    """دالة لحساب إجمالي سنوات الخبرة بدقة"""
    starts = [parse_date_any(x) for x in to_list(starts_cell)]
    ends   = [parse_date_any(x) for x in to_list(ends_cell)]

    n = min(len(starts), len(ends))
    if n == 0:
        return 0.0

    total_days = 0
    for i in range(n):
        s = starts[i]
        e = ends[i]
        if pd.isna(s):
            continue
        if pd.isna(e):
            e = pd.Timestamp.today().normalize()
        if e < s:
            s, e = e, s
        total_days += (e - s).days

    return round(total_days / 365.25, 2)

# تطبيق الدالة على كامل البيانات لإنشاء عمود سنوات الخبرة الرقمي
df["years_experience"] = df.apply(lambda r: years_experience_from_lists(r["start_dates"], r["end_dates"]), axis=1)

print("✅ تم حساب سنوات الخبرة.")
display(df[['start_dates', 'end_dates', 'years_experience']].head(2))
```

```python
import numpy as np
import re

def extract_required_years(text):
    """دالة لاستخراج رقم الخبرة المطلوبة من وسط النص"""
    if pd.isna(text):
        return 0.0
    text = str(text)
    nums = re.findall(r"\d+", text)
    if not nums:
        return 0.0
    return float(nums[0])

df["required_years"] = df["experiencere_requirement"].apply(extract_required_years)
df["skills_required"] = df["skills_required"].fillna("")

keep_cols = [
    "edu_preview", "skills_preview", "years_experience",
    "educationaL_requirements", "required_years", "skills_required",
    "matched_score"
]
df_final = df[keep_cols].copy()

print(f"✅ تم تجهيز البيانات النهائية بنجاح! حجم البيانات: {df_final.shape}")
display(df_final.head(17))
```

```python
from sentence_transformers import SentenceTransformer
import pandas as pd

model_sbert = SentenceTransformer('all-MiniLM-L6-v2')

df_final['text_for_embedding'] = df_final['edu_preview'] + " " + df_final['skills_preview']

X_text_vectors = model_sbert.encode(df_final['text_for_embedding'].tolist(), show_progress_bar=True)

df_vectors = pd.DataFrame(X_text_vectors)

print("✅ تم إنشاء df_vectors بنجاح!")
```

```python
import pandas as pd

X = pd.concat([
    df_vectors.reset_index(drop=True),
    df_final[['years_experience', 'required_years']].reset_index(drop=True)
], axis=1)

y = df_final['matched_score'].reset_index(drop=True)

X.columns = [str(i) for i in X.columns]

X = X.fillna(0)
y = y.fillna(0)

print(f"✅ المصفوفة X جاهزة الآن لمشروعك.")
print(f"📊 حجم البيانات: {X.shape[0]} سيرة ذاتية | عدد الميزات: {X.shape[1]}")

display(X.head(2))
```

```python
import numpy as np

# دالة لتحويل الرقم إلى فئة
def categorize_score(score):
    if score < 0.5:
        return 0  # ضعيف
    elif score <= 0.75:
        return 1  # متوسط
    else:
        return 2  # ممتاز

y_class = df_final['matched_score'].apply(categorize_score)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y_class, test_size=0.2, random_state=42)
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report

model_class = XGBClassifier(
    n_estimators=500,
    learning_rate=0.05,
    max_depth=6,
    objective='multi:softprob', # مخصص للتصنيف المتعدد
    random_state=42
)

# التدريب
print("⏳ جاري تدريب موديل التصنيف...")
model_class.fit(X_train, y_train)
print("✅ انتهى التدريب!")
```

```python
# التوقع
y_pred = model_class.predict(X_test)

# حساب الدقة
accuracy = accuracy_score(y_test, y_pred)
print(f"\n🎯 دقة الموديل في التصنيف: {accuracy:.2%}")

# تقرير تفصيلي لكل فئة
print("\n--- تقرير الأداء التفصيلي ---")
print(classification_report(y_test, y_pred, target_names=['ضعيف', 'متوسط', 'ممتاز']))
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
import matplotlib.pyplot as plt


candidate_text = df_final['edu_preview'] + " " + df_final['skills_preview']
X_text_vectors_filtered = model_sbert.encode(candidate_text.tolist(), show_progress_bar=False)

job_text = df_final['educationaL_requirements'] + " " + df_final['skills_required']
job_vectors = model_sbert.encode(job_text.tolist(), show_progress_bar=False)

semantic_sim = [cosine_similarity(X_text_vectors_filtered[i].reshape(1, -1),
                                  job_vectors[i].reshape(1, -1))[0][0]
                for i in range(len(X_text_vectors_filtered))]

X_final = pd.concat([
    pd.DataFrame(X_text_vectors_filtered).reset_index(drop=True),
    pd.Series(semantic_sim, name='semantic_similarity').reset_index(drop=True),
    df_final['years_experience'].reset_index(drop=True),
    df_final['required_years'].reset_index(drop=True),
    (df_final['years_experience'] - df_final['required_years']).rename('exp_gap').reset_index(drop=True)
], axis=1)

X_final.columns = [str(i) for i in X_final.columns]

def get_label(score):
    if score < 0.5: return 0
    elif score <= 0.75: return 1
    else: return 2

y_class = df_final['matched_score'].apply(get_label).reset_index(drop=True)

print(f"✅ تم تجهيز البيانات بنجاح! حجم X: {X_final.shape} | حجم y: {len(y_class)}")

X_train, X_test, y_train, y_test = train_test_split(X_final, y_class, test_size=0.2, random_state=42)

final_model = XGBClassifier(
    n_estimators=1000,
    learning_rate=0.03,
    max_depth=8,
    objective='multi:softprob',
    random_state=42,
    eval_metric="merror")

eval_set = [(X_train, y_train), (X_test, y_test)]


final_model.fit(
    X_train, y_train,
    eval_set=eval_set,
    verbose=False
)

y_pred = final_model.predict(X_test)

print("\n" + "="*30)
print(f"🎯 الدقة النهائية للمشروع: {accuracy_score(y_test, y_pred):.2%}")
print("="*30)
print("\n--- تقرير الأداء التفصيلي ---")
print(classification_report(y_test, y_pred, target_names=['ضعيف', 'متوسط', 'ممتاز']))

# --- الخطوة 4: رسم المخطط ---
results = final_model.evals_result()
epochs = len(results['validation_0']['merror'])
x_axis = range(0, epochs)

train_accuracy = [1 - x for x in results['validation_0']['merror']]
test_accuracy = [1 - x for x in results['validation_1']['merror']]

plt.figure(figsize=(10, 6))
plt.plot(x_axis, train_accuracy, label='Training Accuracy (دقة التدريب)', color='blue', linewidth=2)
plt.plot(x_axis, test_accuracy, label='Testing Accuracy (دقة الاختبار)', color='orange', linewidth=2)

plt.title('XGBoost Learning Curve (مخطط الأوفر فيتينغ)', fontsize=14, fontweight='bold')
plt.xlabel('Number of Trees (عدد الأشجار)', fontsize=12)
plt.ylabel('Accuracy (الدقة)', fontsize=12)
plt.legend(loc='lower right', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
import pandas as pd

X_smart = pd.DataFrame({
    'sim': semantic_sim,
    'exp': df_final['years_experience'].values,
    'req': df_final['required_years'].values,
    'gap': (df_final['years_experience'] - df_final['required_years']).values
})

y_binary = df_final['matched_score'].apply(lambda score: 0 if score < 0.6 else 1).reset_index(drop=True)

X_train, X_test, y_train, y_test = train_test_split(X_smart, y_binary, test_size=0.2, random_state=42)

# حساب نسبة الانحياز لمعالجتها
neg_count = len(y_train[y_train == 0])
pos_count = len(y_train[y_train == 1])
imbalance_ratio = neg_count / pos_count  # توازن الكفة


model_improved = XGBClassifier(
    n_estimators=200,        # زيادة طفيفة في الأشجار
    learning_rate=0.02,      # إبطاء التعلم ليصبح أدق
    max_depth=3,             # أشجار سطحية لمنع الحفظ
    subsample=0.8,           # تدريب على 80% فقط عشوائياً
    colsample_bytree=0.8,    # استخدام بعض الميزات عشوائياً
    gamma=1,                 # منع التفرع العشوائي
    scale_pos_weight=imbalance_ratio, # حل مشكلة انحياز البيانات
    random_state=42,
    eval_metric="error"
)

eval_set = [(X_train, y_train), (X_test, y_test)]
model_improved.fit(X_train, y_train, eval_set=eval_set, verbose=False)

y_pred = model_improved.predict(X_test)

print("\n" + "★"*40)
print(f"🎯 الدقة بعد التحسين الاحترافي: {accuracy_score(y_test, y_pred):.2%}")
print("★"*40)
print("\n--- التقرير التفصيلي ---")
print(classification_report(y_test, y_pred, target_names=['0: غير مناسب', '1: مناسب']))

# 6. رسم المخطط المحسن
results = model_improved.evals_result()
x_axis = range(0, len(results['validation_0']['error']))

plt.figure(figsize=(10, 5))
plt.plot(x_axis, [1-x for x in results['validation_0']['error']], label='Training Accuracy', color='blue', linewidth=2)
plt.plot(x_axis, [1-x for x in results['validation_1']['error']], label='Testing Accuracy ', color='orange', linewidth=2)

plt.title('Optimized Learning Curve (High Accuracy & Stable)', fontsize=14, fontweight='bold')
plt.xlabel('Number of Trees', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.legend(loc='lower right', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
